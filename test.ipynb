```python
import numpy as np
```

```python
def readFile():
    file_name = "cord.txt"
    with open(file_name, 'r') as file:
        lines = file.readlines()
    array_2d = [list(map(int, line.strip().replace(
        '(', '').replace(')', '').split(','))) for line in lines]
    return array_2d
```

```python
readFile()
```

```python
cord=np.array(readFile())
```
